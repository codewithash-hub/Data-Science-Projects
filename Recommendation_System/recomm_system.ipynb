```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

# Prepare Your Data

```python
data = pd.read_csv('shopping_trends.csv')
data.head()
```

```python
data_dict = {
    'UserID': data['Customer ID'],
    'ItemPurchased': data['Item Purchased'],
    'Rating': data['Review Rating']
}

df = pd.DataFrame(data_dict)
df
```

# Create the Recommendation System

```python
# create a user item matrix
user_item_matrix = df.pivot_table(index = 'UserID', columns = 'ItemPurchased', values = 'Rating').fillna(0)

# calculate similarity between users cosine similarity
user_similatiry = cosine_similarity(user_item_matrix)

# Get similar users
def similar_users(user_id, metrix):

    user = metrix.iloc[user_id - 1].values.reshape(-1, 1)
    sim_scores = list(enumerate(cosine_similarity(user, metrix)[0]))
    sim_scores = sorted(sim_scores, key = lambda x: x[1], reverse = True)
    sim_scores = sim_scores[1:]
    user_indices = [i[0] for i in sim_scores]

    return user_indices

# Get similar users to UserID 1
similar_to_User = similar_users(user_item_matrix, user_similatiry)
print(f"User similar to User 1: {similar_to_User}")
```
